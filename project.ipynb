```python
# import python Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import model from sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
# impport label encoder for converting word into numeric value like 1 or 0
from sklearn.preprocessing import LabelEncoder

# intialing the LabelEncoder
le = LabelEncoder()
```

```python
# Preprocessing Encode categorical columns
df['PlacementStatus'] = le.fit_transform(df['PlacementStatus'])   # for e.g. converting 'NotPlaced' to 0 and 'Placed' to 1

df['ExtracurricularActivities'] = le.fit_transform(df['ExtracurricularActivities'])       # converting 'NO' to 0 and 'YES'to 1

df['PlacementTraining'] = le.fit_transform(df['PlacementTraining']) # converting 'No' to 0 and 'YES' to 1

```

```python
# displaying new data after doing change in old data
display(df)
```

```python
sns.barplot(x='ExtracurricularActivities', y='PlacementStatus', data=df)
plt.show()
```

```python
# checking null value
df.isna().sum()
```

```python
# X = Independent and y = Dependent
X= df.drop('PlacementStatus', axis = 1)
y = df['PlacementStatus']
```

```python
display(X)
```

```python
display(y)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 1)
```

```python

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python

y_pred = lr.predict(X_test)
y_pred
```

```python

from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
```

```python

accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
# checking value which is 'ACTUAL POSITIVE' and 'ACTUAL NEGATIVE' model predict them 'POSITIVE and NEGATIVE'
conf_matrix= confusion_matrix(y_test, y_pred)
conf_matrix_df = pd.DataFrame(conf_matrix, index=['Actual Positive','Actual Negative'], columns=['Predicted Postive','Predicted Negative'])

print(conf_matrix_df)
```
